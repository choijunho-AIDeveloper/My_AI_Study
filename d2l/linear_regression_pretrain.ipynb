```python
import pandas as pd

house_data = "house_price.csv"

data = pd.read_csv(f"./{house_data}")
print(data.head())
print(data.shape)
```

```python
from sklearn.model_selection import train_test_split

# x: 'MEDV'열을 제외한 csv data를 출력, y: 'MEDV'열 데이터를 출력
x = data.drop(['MEDV'], axis = 1)
y = data['MEDV']


x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
print(x_train.head)
print(f"x_train_shape : {x_train.shape}")
print(y_train.head)
print(f"y_train_shape : {y_train.shape}")
```

```python
from sklearn.linear_model import LinearRegression

regressor = LinearRegression()
regressor.fit(x_train, y_train)

y_pred = regressor.predict(x_test)

print(y_pred[:5])
print(y_test[:5])
```

```python
from sklearn.model_selection import train_test_split

# x: 'MEDV'열을 제외한 csv data를 출력, y: 'MEDV'열 데이터를 출력
x = data.drop(['MEDV'], axis = 1)
y = data['MEDV']


x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
print(x_train.head)
print(f"x_train_shape : {x_train.shape}")
print(y_train.head)
print(f"y_train_shape : {y_train.shape}")
```

```python
from sklearn.linear_model import LinearRegression

regressor = LinearRegression()
regressor.fit(x_train, y_train)

y_pred = regressor.predict(x_test)

print(y_pred[:5])
print(y_test[:5])
```

```python
import matplotlib.pyplot as plt
import numpy as np

colors = np.random.randn(102)

plt.scatter(y_test, y_pred, c = colors)
plt.xlabel("price")
plt.ylabel("pred")
plt.title("real vs pred")
plt.show()
```

```python
from sklearn.metrics import mean_squared_error

MSE = mean_squared_error(y_test, y_pred)
result = 100 - MSE
print(f"정확도 : {result}")
```

```python
y_result = y_test - y_pred
x_result = np.arange(len(y_result))


plt.scatter(x_result, y_result, color = "green")
plt.ylim(-30, 30)
plt.axhline(y = 0, color = "yellow", linestyle = "dashed")
plt.ylabel("Test - Pred")
plt.show()
```
